```python
# we get the labels of each photo by the xml files which has the same name as the photo. And we use the box data from the xml file to obtain the face location and crop it out.
import json
import shutil,os
from xml.dom import minidom
def dis(datadir,maskDir,noMaskDir):
    path = original_dataset_dir+datadir #position of the folder where xml andjpg files are
    filelist = os.listdir(path)
    facelist=[]      #put picture 
    masklist=[]
    for file in filelist:
        if ( ".xml" in file ):
            picture1 = file.split(".",1) #use“.”to divide the file name
            picture=picture1[0]   #get the file name
            compPath=os.path.join(path, file)
            domxml = minidom.parse(compPath)
            names = domxml.getElementsByTagName('name')
            xmin = domxml.getElementsByTagName('xmin')[0].firstChild.data
            xmax = domxml.getElementsByTagName('xmax')[0].firstChild.data
            ymin = domxml.getElementsByTagName('ymin')[0].firstChild.data
            ymax = domxml.getElementsByTagName('ymax')[0].firstChild.data
            imageName = picture+".jpg"
            image=Image.open(os.path.join(path, imageName))
            image = image.crop((int(xmin), int(ymin), int(xmax), int(ymax)))
            image = image.convert('RGB')
            try:
                if names[0].firstChild.data=='face_mask':
                    image.save(os.path.join(maskDir, imageName))
                else:
                    image.save(os.path.join(noMaskDir, imageName))
            except:
                print(imageName)
```

```python
batch_size = 25 #10
TRAINING_DIR = train_dir
train_datagen = ImageDataGenerator(rotation_range=15,
                                    rescale=1./255,
                                    shear_range=0.1,
                                    zoom_range=0.2,
                                    horizontal_flip=True,
                                    width_shift_range=0.1,
                                    height_shift_range=0.1)#rescale=1.0/255.
train_generator = train_datagen.flow_from_directory(TRAINING_DIR,
                                                    batch_size=batch_size,
                                                    class_mode='binary',
                                                    target_size=(150, 150))

VALIDATION_DIR = validation_dir
validation_datagen = ImageDataGenerator(rescale=1.0/255)#.
validation_generator = validation_datagen.flow_from_directory(VALIDATION_DIR,
                                                              batch_size=batch_size,
                                                              class_mode='binary',
                                                              target_size=(150, 150))
```

```python
# See how our generator work. It would randomly rescale, reotate, zoom, shear, shift, flip the photo according to the settings we defined previously to produce more samples to adapt the changeable environment.
plt.figure(figsize=(12, 8))
for i in range(0, 15):
    plt.subplot(3,5, i+1)
    plt.imshow(train_generator[0][0][0])
plt.tight_layout()
plt.show()
```

```python
# we set the early stop to help us break the training process. When there are 10 epochs with no improvement, the training will be stopped. Also we monitors the validation accuracy and if no improvement is seen for 2 epochs, the learning rate will be reduced to half of it (no less than 1e-5).
import tensorflow as tf
# from matplotlib import pyplot
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
earlystop = EarlyStopping(patience=10)
learning_rate_reduction = ReduceLROnPlateau(monitor='val_acc', 
                                            patience=2, 
                                            verbose=1, 
                                            factor=0.5, 
                                            min_lr=0.00001)
callbacks = [earlystop, learning_rate_reduction]
```

```python
def fit_model(lr,axes):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
        tf.keras.layers.MaxPooling2D(2, 2),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2, 2),
#         tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
#         tf.keras.layers.MaxPooling2D(2, 2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer=RMSprop(lr=lr), loss='binary_crossentropy', metrics=['acc'])
    # fit model
    # in order to speed up the process of choosing hyperparameter lr, we set other parameters to speed up the process but not affect the result much.
    history = model.fit(
      train_generator,
      steps_per_epoch=100,  #total_train_num//batch_size,
      epochs==15,
      validation_data=validation_generator,
      validation_steps=40,  #total_val_num//batch_size,
      callbacks=callbacks
      )
    # plot learning curves
    axes.plot(history.history['acc'],'r',label='Training accuracy')
    axes.plot(history.history['val_acc'],'b',label='Validation Accuracy')
    title ='learning rate ='+str(lr)
    axes.set_title(title)
    axes.legend()
    
```

```python
# from the results we get that learning rate = 0.01 get bad result which may due to underfiting. On the other hand, when it eauqls to 1e-5, the accuracy would have fluctuations, which may be caused by overfitting. Since when seting learning rate to 1e-4 and 1e-3 have the similar result, we finally decide set it to 1e-3, which processes faster and more unlikely to fail into local optimality.
lr=[0.00001,0.0001,0.001,0.01]
fig,axes=plt.subplots(2,2,figsize=(12,7))
for i in range(len(lr)):
    fit_model(lr[i],axes[i//2,i%2])
# show learning curves

plt.show()
```

```python
import cv2
from keras.preprocessing import image
from keras.models import load_model
import numpy as np
import dlib
from PIL import Image
model = load_model('./maskmodel.h5')
# detector = dlib.get_frontal_face_detector()
classfier = cv2.CascadeClassifier("D:\\language\\Anaconda3\\envs\\deep_learning\\Library\\etc\\haarcascades\\haarcascade_frontalface_alt2.xml")

video=cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_SIMPLEX
def rec(img):
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faceRects = classfier.detectMultiScale(gray, scaleFactor = 1.2, minNeighbors = 3, minSize = (32, 32))
    # dets=detector(gray,1)
    if len(faceRects) > 0:            #大于0则检测到人脸                                   
        for faceRect in faceRects:  #单独框出每一张人脸
            x, y, w, h = faceRect
            left,top,right,bottom = x - 10, y - 10, x + w + 10, y + h + 10 
            # left=face.left()
            # top=face.top()
            # right=face.right()
            # bottom=face.bottom()
            cv2.rectangle(img,(left,top),(right,bottom),(0,255,255),2)
            img1=cv2.resize(img[top:bottom,left:right],dsize=(150,150))
            img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
            img1 = np.array(img1)/255.
            img_tensor = img1.reshape(-1,150,150,3)
            prediction =model.predict(img_tensor)    
            if prediction[0][0]>0.5:
                result='no-mask'
            else:
                result='have-mask'
            cv2.putText(img, result, (left,top), font, 2,(255, 255, 100), 2, cv2.LINE_AA)
        cv2.imshow('Video', img)
while video.isOpened():
    res, img_rd = video.read()
    if not res:
        break
    rec(img_rd)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
